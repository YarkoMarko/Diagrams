# Завдання 3
Очистіть дані від викидів. Виведіть очищений датафрейм

```python
for column in df.select_dtypes("number"):
  df[column].plot(kind="box", vert=False, figsize=(14, 6))
  plt.show()
```

```python
def remove(column, df):
  column_ = df[column]

  q1 = column_.quantile(q=0.25)
  q3 = column_.quantile(q=0.75)
  iqr = q3 - q1

  lower = q1 - iqr * 1.5
  upper = q3 + iqr * 1.5

  mask = (column_ >= lower) & (column_ <= upper)

  return df[mask]

for column in df.select_dtypes(include="number").columns:
  df = remove(column, df)
```

```python
for column in df.select_dtypes("number"):
  df[column].plot(kind="box", vert=False, figsize=(14, 6))
  plt.show()
```

# Завдання 4
Намалюйте матрицб кореляцій. Для найбільш корельованих ознак намалюйте точкову діаграму.

```python
columns_n = df.select_dtypes(include="number").columns
corr = df[columns_n].corr(method="spearman")
fig = plt.figure(figsize=(8, 8))
plt.matshow(corr, cmap="RdBu", fignum=fig.number)
plt.xticks(range(len(corr.columns)), corr.columns, rotation="vertical")
plt.yticks(range(len(corr.columns)), corr.columns)
plt.colorbar()
```
